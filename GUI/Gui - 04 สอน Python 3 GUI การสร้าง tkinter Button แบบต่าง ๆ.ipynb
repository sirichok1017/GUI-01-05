## รูปแบบการเขียน tkinter Button

* Button(root, text="ตัวอักษร",fg="สี", bg="สี"",bd=10,padx=30).pack(fill=X)


* text - ตัวอักษรที่ต้องการเขียน
* fg - foreground ต้องการใส่สีใน foreground
* bg - background ต้องการใส่สีใน background
* bd - เป็นการกำหนดขอบ Button
* pad - ขนาด Button
* .pack() - กำหนดตัวอักษรให้ชิดซ้าย ชิดขวา หรือตรงกลาง
    * .pack(anchor=W) - ต้องการให้ตัวอักษรชิดซ้าย
    * .pack(anchor=E) - ต้องการให้ตัวอักษรชิดขวา
    * .pack(fill=X) - ต้องการให้ตัวอักษรอยู่ตรงกลาง

```python
from tkinter import *  #โหลดโมดูล tkinter 

#สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root = tk.Tk()  

#สร้างออบเจ็คปุ่ม Button widget
Button(root, text="phone").pack()

#คำสั่งแสดงผล GUI
root.mainloop()   

#ถ้า Run จะได้ widget ที่มี Button คำว่า "phone"
```

```python
from tkinter import *  #โหลดโมดูล tkinter 

#สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root = tk.Tk()  

#Font ทั้งหมด widget เป็น consolas ขนาด 20
root.option_add("*Font", "consolas 20")

#สร้างออบเจ็ค ที่เป็นรูปภาพ
photo = PhotoImage(file="Gui03\yinyang.png")

#อ้างอิงชื่อออบเจ็คและเพิ่มออปชัน image สำหรับการแสดงผล
#เพิ่มระยะห่างตัวอักษร กับรูปใน Button เดียวกัน โดยการเพิ่ม padx=20
#สามารถเพิ่มระยะห่างตัวอักษร กับรูป ได้ทั้งแนวตั้ง และแนวนอน
Button(root, text="Yin Yang", image=photo, compound=TOP).pack()
Button(root, text="Yin Yang", image=photo, compound=TOP, pady=20).pack()
Button(root, text="Yin Yang", image=photo, compound=BOTTOM).pack()

#คำสั่งแสดงผล GUI
root.mainloop()   

#ถ้า Run เห็นได้ว่ามีรูป "Yin Yang" อันที่ 2 ตัวอักษร กับรูปมีระยะห่างมากกว่า Button อื่น
```
