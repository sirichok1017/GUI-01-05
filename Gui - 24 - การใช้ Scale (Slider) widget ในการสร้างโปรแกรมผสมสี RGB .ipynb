```python
## การคำนวณโปรแกรมผสมสี RGB ตอนที่ 2 โดยการเขียน code
from tkinter import *

def on_drag(e):                                                                 # สร้าง event
        color_hex = f'#{sc[0].get():02X}{sc[1].get():02X}{sc[2].get():02X}'
        tv_color.set(color_hex)                                                 # เป็นการกำหนดค่าของ Code สี
        lbl_color["bg"] = color_hex                                             # เป็นการกำหนดสีของ background บรรทัด code สี

root = Tk()
root.option_add("*Font", "consolas 20")
tv_color = StringVar()
rgb = ['red', 'green', 'blue']                  # กำหนด list
sc = []                                         
for i ,c in enumerate(rgb):                     # เป็นการสร้าง loop โดยการเข้าถึง rgb ทีละตัว
        Label(root, text=c, fg=c).grid(row=i, column=0, sticky="sw")
    # Code บรรทัดบน เป็นการเขียนแทน 3 บรรทัดล่างนี้
    # Label(root, text="red", fg="red").grid(row=0, column=0, sticky="sw")
    # Label(root, text="green", fg="green").grid(row=1, column=0, sticky="sw")
    # Label(root, text="blue", fg="blue").grid(row=2, column=0, sticky="sw")
        w = Scale(root, from_=0, to=255, orient=HORIZONTAL, length=200, width=30, fg=c) # w จะเป็นตัวแทน red green blue
        w.grid(row=i, column=1)                 # กำหนดว่าจะให้ rows and column เป็นเท่าไร
        w.set(128)                              # กำหนดให้ตั้งค่าเริ่มต้น คือ 128
        w.bind('<B1-Motion>', on_drag)          # เป็นการกำหนด ถ้าเลื่อนปุ่ม slider ค่าที่ได้จะเปลี่ยนไป
        w.bind('<Button-1>', on_drag)           # เป็นการกำหนด ถ้าคลิก แล้วค่าที่ได้จะเปลี่ยนไป
        sc.append(w)                            # sc จะเก็บ object ที่เป็นตัว scale ไว้ 3 ตัว

    # r = Scale(root, from_=0, to=255, orient=HORIZONTAL, length=200, width=30, fg="red")
    # r.grid(row=0, column=1)
    # r.set(128)
    # r.bind('<B1-Motion>', on_drag)
    # r.bind('<Button-1>', on_drag)

    # g = Scale(root, from_=0, to=255, orient=HORIZONTAL, length=200, width=30, fg="green")
    # g.grid(row=1, column=1)
    # g.set(128)
    # g.bind('<B1-Motion>', on_drag)
    # g.bind('<Button-1>', on_drag)

    # b = Scale(root, from_=0, to=255, orient=HORIZONTAL, length=200, width=30, fg="blue")
    # b.grid(row=2, column=1)
    # b.set(128)
    # b.bind('<B1-Motion>', on_drag)
    # b.bnd('<Button-1>', on_drag)
lbl_color = Label(root, textvariable=tv_color)
lbl_color.grid(row=3, columnspan=2, sticky="news")
root.mainloop()
```

```python
# ตาราง rating score (Example)
from tkinter import *

def rating():                                                   # ตัวอย่างการสร้างโปรแกรม rating
    def on_drag(e):
        total = 0
        for score in sc:
            total += score.get()
        rating_score.set(f'avg. = {total / len(sc):.2f}')       # คำสั่งนี้เป็นการคำนวนค่า average ของคะแนนที่เราเลือก
        pass

    root = Tk()
    root.option_add("*Font", "consolas 20")                     # กำหนด Font ให้เป็น consolas size 20 
    rating_score = StringVar()
    criteria = ['price', 'performance', 'design', 'service', 'reliability'] # กำหนด list กี่ตัวก็ได้
    sc = []
    for i ,c in enumerate(criteria):
            Label(root, text=c).grid(row=i, column=0, sticky="sw")
            w = Scale(root, from_=1, to=10, orient=HORIZONTAL, length=200, width=30)    # กำหนดคะแนนให้มีค่า 1 ถึง 10
            w.grid(row=i, column=1)                             # กำหนดว่าจะให้ rows and column เป็นเท่าไร
            w.set(5)                                            # กำหนดค่าเริ่มต้น = 5
            w.bind('<B1-Motion>', on_drag)                      # เป็นการกำหนด ถ้าเลื่อนปุ่ม slider ค่าที่ได้จะเปลี่ยนไป
            w.bind('<Button-1>', on_drag)                       # เป็นการกำหนด ถ้าคลิก แล้วค่าที่ได้จะเปลี่ยนไป
            sc.append(w)

    lbl_color = Label(root, textvariable=rating_score, bg = "gold")     # กำหนดให้ tab average เป็นสีทอง
    lbl_color.grid(columnspan=2, sticky="news")                
    root.mainloop()

if __name__ == '__main__':
    rating()
```
