## รู้จักโมดูล tkinter , ttk 

ชื่อ tkinter มาจาก TK interface หรือเรียกสั้นๆ ว่า TK ได้รับการพัฒนาโดย Fredrik Lundn เป็นโมดูลมาตรฐานของภาษาไพธอนที่ได้รับความนิยมที่ใช้สำหรับพัฒนาโปรแกรมแบบ GUI ทำงานได้ทั้งบน ระบบปฏิบัติการ Windows, MacOS หรือ Linux โมดูล tkinter เป็นการรวมภาษาไพธอนเข้ากับ TK GUI toolkit ซึ่งเป็นซอฟต์แวร์ที่นำมาใช้งานได้ฟรี ที่ประกอบด้วย widget ต่างๆ สำหรับการสร้างโปรแกรมแบบ GUI จำนวนมาก เช่น Frame, Menu, Text field, Label เป็นต้น

TK interface จะทำให้เราสามารถเขียนโปรแกรมที่มีส่วนติดต่อกับผู้ใช้ที่เป็นแบบกราฟิกหรือว่า GUI User Interface ได้ โดยการเขียนเป็นโค้ชที่เขียนด้วย python ตอนที่ output ออกมานะครับจะเป็นแบบกราฟฟิก

```python
import tkinter as tk  #โหลดโมดูล tkinter เข้ามายังโปรแกรมทุกคลาสตั้งชื่อเล่นว่า TK

root = tk.Tk()  #สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root.mainloop()   #คำสั่งแสดงผล GUI

#ถ้า Run จะได้ Windows ขึ้นมาอันนึง
```

```python
import tkinter as tk  #โหลดโมดูล tkinter ชื่อเล่นว่า TK

root = tk.Tk()  #สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
tk.Label(root, text="hello world").grid() #สร้าง Label widget
root.mainloop()  #คำสั่งแสดงผล GUI

#ถ้า Run เราจะเห็นว่าจะมี text เพิ่มเข้ามาว่า "hello world"
```

```python
import tkinter as tk  #โหลดโมดูล tkinter ชื่อเล่นว่า TK

root = tk.Tk()  #สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root.option_add("*Font", "consolas 20")  #Font ทั้งหมด widget เป็น consolas ขนาด 20
tk.Label(root, text="hello world").grid() #สร้าง Label widget
root.mainloop() #คำสั่งแสดงผล GUI

#ถ้า Run เราจะเห็นว่ามีการ เปลี่ยน Font และเพิ่มขนาดของ text ให้ใหญ่ขึ้น
```

```python
import tkinter as tk  #โหลดโมดูล tkinter ชื่อเล่นว่า TK

root = tk.Tk()  #สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root.option_add("*Font", "consolas 20")  #Font ทั้งหมด widget เป็น consolas ขนาด 20
root.title("I love tkinter very much")  #แสดงข้อความที่ title bar "I love tkinter very much"
tk.Label(root, text="hello world").grid() #สร้าง Label widget
root.mainloop()  #คำสั่งแสดงผล GUI

#ถ้า Run เราจะเห็นว่า หัวของ Windows หรือ title bar จะมีคำว่า "I love tkinter very much" เพิ่มขึ้นมา
```

```python
import tkinter as tk  #โหลดโมดูล tkinter ชื่อเล่นว่า TK

root = tk.Tk()  #สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root.option_add("*Font", "consolas 20")  #Font ทั้งหมด widget เป็น consolas ขนาด 20
root.title("I love tkinter very much")  #แสดงข้อความที่ title bar "I love tkinter very much"
for i in range(10):  #ทำ for Loop 10 ครั้ง
    tk.Label(root, text="hello world").grid() #สร้าง Label widget
root.mainloop()  #คำสั่งแสดงผล GUI

#ถ้า Run เราจะเห็นว่าจะมี text เพิ่มเข้ามาว่า "hello world" 10 คำ
```
