## การจัดวางตำแหน่ง widgets ด้วยเมธอด pack()

    เป็นรูปแบบการจัดวางตำแหน่ง widget ลงบนหน้าต่าง Frame โดยอาศัยความสัมพันธ์ระหว่าง 2 widget และใช้ออปชันกำหนดค่าให้กับเมธอด ซึ่งจะเข้าไปจัดการตำแหน่งให้กับ widget อีกครั้ง มีโครงสร้าง รูปแบบการเรียกใช้งานเมธอด pack() ดังนี้ (ค่าเริ่มต้น widget จะเรียงจากบนลงล่าง)


* w.pack (option=value, ...)

    * w - widget ที่ต้องการจัดตำแหน่ง
    * pack() - เมธอดการจัดตำแหน่ง 
    * option=value - คุณสมบัติการจัดตำแหน่ง widget

```python
from tkinter import * #โหลดโมดูล tkinter

#สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root = Tk()

#method 1:การสร้าง widgets ขึ้นมา
Button(root, text="apple").pack(side=LEFT)

#คำสั่งแสดงผล GUI
root.mainloop() 

#ถ้า Run จะเห็น widgets "apple" ขึ้นมา ชิดทางซ้าย
```

```python
from tkinter import * #โหลดโมดูล tkinter

#สร้างออบเจ็ค ชื่อ root (ชื่ออะไรก็ได้)
root = Tk()

#Font ทั้งหมด widget เป็น consolas ขนาด 25
root.option_add("*Font", "impact 25")


#การสร้าง widgets คำว่า "rainbow" ทีละตัวอักษร
for c in "rainbow":
        Button(root, text=c).pack(side=LEFT)

#คำสั่งแสดงผล GUI
root.mainloop() 

#ถ้า Run จะเห็นได้ว่า widget "rainbow" ขึ้นมา สามารถกดได้ทีละตัวอักษร
#โดยการเรียง "rainbow" จากซ้ายไปขวา
```
